***Zadanie 1.***  
Zdefiniuj w notatniku funkcję $f(x)= x^{3} - 3x$ i sporządź jej wykres dla argumentów z przedziału:  
1. $x\in<-1,1>$  
2. $x\in<-5,5>$  
3. $x\in<0,5>$
  
***Wskazówki***
Jako argumentu funkcji można użyć numpy array.  

Do wizualizacji wyników w Pythonie używa się pakietu [Matplotlib](https://matplotlib.org/). Działanie tego pakietu prezentuje przykład:

```python
x1=np.linspace(-1,1)
x2=np.linspace(-5,5)
x3=np.linspace(0,5)
y=np.array(x1**3-3*x1)
plt.plot(x1,y,label='x^3-3x')
plt.title("Wykres funkcji dla przedziału -1,1")
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.grid()
plt.show()
y2 = np.array(x2**3-3*x2)
plt.plot(x2,y2,label='x^3-3x')
plt.title("Wykres funkcji x^3-3x dla przedziału -5,5")
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.grid()
plt.show()
y3= np.array(x3**3-3*x3)
plt.plot(x3,y3)
plt.title("Wykres funkcji x^3-3x dla przedziału 0,5")
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.grid()
plt.show()





```

***Komentarz:
Definiowanie prostej funkcji oraz użycie funkcji linspace do określenia przedziału argumentów. Za pomocą biblioteki plt możemy wyrysować wykres, nadać mu tytuł, podpisać osie.***

Wspierając się dokumentacją [Matplotlib](https://matplotlib.org/) dodaj do wykresu etykiety osi, tytuł, grida i legendę.

***Zadanie 2.***  
Dla funkcji z zadania pierwszego używając wektora $x \in <-10,10>$ wyrysuj wykres dla argumentów:
1. $x\in<-1,1>$  
2. $x\in<-10,-1>$  
3. $x\in<1,10>$

Używając do tego funkcji: [ylim](https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.ylim.html), [xlim](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.xlim.html), [axis](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.axis.html)

```python
def funkcja(x):
    return x**3-3*x
x =np.linspace(-10,10,20)
plt.plot(x,funkcja(x))
plt.xlim((-1,1))

plt.title("Wykres dla przedziału -1;1")
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.grid()
plt.show()

plt.plot(x,funkcja(x))
plt.title("Wykres dla przedziału -10,1")
plt.xlim(-10,-1)
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.grid()
plt.show()

plt.plot(x,funkcja(x))
plt.title("Wykres dla przedziału 1,10")
plt.xlim(1,10)
plt.xlabel("Oś x")
plt.ylabel("Oś y")
plt.grid()
plt.show()


```

***Komentarz:Zapoznanie się z funkcją xlim, która zawęża nam wykres w określonej dziedzinie iksów od x_min do x_max. Analogicznie możemy z funkcją ylim, dla określonych y.***

***Zadanie 3.*** 
Oblicz ilość ciepła, które wydzieli się podczas hamowania jeżeli opisuje je zależność:  
  
$Q = \cfrac{mv^2}{2}$  
  
jeżeli $m = 2500g$, $v = 60km/h$, wypisz wynik oraz podaj w kilokaloriach i dżulach. Wykreśl wykresy które zobrazują zmianę ilości ciepła w procesie hamowania dla ciała o masie $3000g$ i prędkości $v\in<200,0>km/h$, przedstawiając ciepło hamowania na pierwszym w skali liniowej na drugim w skali logarytmicznej.  

***Wsakzówka:*** Do stworzenia wykresów z osią logarytmiczną skorzystaj z funkcji: [semilogy](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.semilogy.html), [xlim](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xlim.html?highlight=xlim#matplotlib.pyplot.xlim) i [figure](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html?highlight=figure#matplotlib.pyplot.figure) z pakietu *Matplotlib*. 

```python
def ilosc_ciepla(m,v):
    return (m*v*v)/2
m = 2500/1000    #zamiana na kilogramy
v = (60*1000)/3600 #zamiana na metry na sekundę
print("Ilość ciepla w dżulach wynosi {:.3f}".format(ilosc_ciepla(m,v)))
print("Ilość ciepla w kilokaloriach wynosi {:.3f}". format(ilosc_ciepla(m,v)*0.0002388459))

speed = np.linspace(200,0)
ilosc_ciepla1 = (m*speed*speed)/2
plt.plot(speed,ilosc_ciepla1)
plt.title("Wykres w skali liniowej")
plt.xlabel("V-prędkość [m/s]")
plt.ylabel("Q-prędkość hamowania [J]")
plt.grid()
plt.show()
plt.title("Wykres w skali logarytmicznej")
plt.xlabel("V-prędkość [m/s]")
plt.ylabel("Q-prędkość hamowania [J]")
plt.grid()
plt.semilogy(speed,ilosc_ciepla1)
plt.show()
```

***Komentarz:Definiowanie prostej funkcji oraz wyrysowanie dla niej dwóch wykresów odpowiednio dla funkcji liniowej oraz logarytmicznej semilogy.***

***Zadanie 4.***  
Częstą praktyką szczególnie w dziedzinie analizy danych, statystyce, uczeniu maszynowym, itp. jest tworzenie na podstawie biblioteki [Matplotlib](https://matplotlib.org/) własnych szablonów wykresów stworzonych na podstawie funkcji. Najprostszy przykład to przeładowanie funkcji plot tak by wykres miał czerwony kolor:

```python
def my_plot(x,y,xlabel,ylabel,title,label):
    if x.shape != y.shape or  min(x.shape)==0:
        return None
    fig, ax = plt.subplots()
    ax.plot(x, y, 'r', label=label)
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
    ax.legend()
    return fig

x = np.linspace(-10,10)
# inicjalizacja wektora f(x) = sin(x)
y = np.sin(x)
my_plot(x,y,'x','y','Funkcja sin', 'sin(x)')
```

***Zadanie 7.***  
Za pomocą funkcji *parallel\_plot*  i przedstaw na jednym z nich [Spirale logarytmiczną](https://pl.wikipedia.org/wiki/Spirala_logarytmiczna) w szerokim przedziale, a w drugim w okolicy zera.

```python
th = np.linspace(0,6*np.pi,90)

# przypadek pierwszy
a=100
b=0.2
r = a*np.exp(b*th)
x1 = r * np.cos(th)
y1 = r * np.sin(th)


# przypadek drugi
a=0.01
b=0.2
r = a*np.exp(b*th)
x2 = r * np.cos(th)
y2 = r * np.sin(th)


x1label = "x"
x2label="x"
y1label="y"
y2label="y"
title='Spirala logarytmiczna'

orientation='-'
main.parallel_plot(x1,y1,x2,y2,x1label,y1label,x2label,y2label,title,orientation)
plt.show()
```

***Komentarz:Za pomocą powyższej funkcji mogę przedstawić spirale logarytmiczne w szerokim przedziale oraz w okolicy zera dobierając odpowiednie współczynniki i przedział kąta.***
